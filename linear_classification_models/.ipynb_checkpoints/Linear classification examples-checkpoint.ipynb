```python
import numpy as np
import pandas as pd
import random
from sklearn.utils import shuffle
import copy
import perceptron
import logistic_regression

def train(epochs, threshold, classifier, train_data):           ##on-line training of data
    best_W = np.zeros(train_data.shape[1])
    best_so_far=train_data.shape[0]
    best_accuracy=0
    for epoch in range(epochs):
        train_data = shuffle(train_data)
        missed=0
        for i in range(train_data.shape[0]):
            miss = classifier.update(train_data.loc[i])
            missed+=miss
        train_accuracy = (train_data.shape[0]-missed)/train_data.shape[0]
        if missed< best_so_far:                             ##save the model with best accuracies
            best_accuracy = train_accuracy
            best_so_far = missed
            best_W = copy.deepcopy(classifier.W)
    print('train_accuracy: ', best_accuracy)
    return best_W
        
def test(classifier, data):                                 ##calculate accuracy on test data
    accuracy = 0
    for i in range(data.shape[0]):
        [p,l] = classifier.predict(data.loc[i])
        if p==l:
            accuracy+=1
    print('accuracy: ', accuracy/data.shape[0])
    return  accuracy/data.shape[0]
```

```python
data = pd.read_csv('./2d_gaussian_data.csv')    ##run classification_data.py to get train and test data
train_data = data[0:4000]
test_data = data[4000:]
test_data = test_data.reset_index()
del test_data['index']
```

```python
import matplotlib.pyplot as plt

c1 = test_data[test_data['class_label']==1]
c2 = test_data[test_data['class_label']==-1]
#print(c1['x1'].values, c1['x2'].values)
plt.plot(c1['x1'].values, c1['x2'].values, 'rx', label='class 1')
plt.plot(c2['x1'].values, c2['x2'].values, 'bx', label='class 0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

```python
##perceptron training and testing

classifier_p = perceptron.perceptron(2,1)
classifier_p.W = copy.deepcopy(train(20, 0, classifier_p, train_data))## train for 20 iterations
print('learnt parameters: ',classifier_p.W)
#test(classifier_p, train_data)
test(classifier_p, test_data)

d1 = train_data[train_data['class_label']==1]
d2 = train_data[train_data['class_label']==-1]

fig,a =  plt.subplots(1,2, figsize=(16,10))
x = np.linspace(min(test_data['x1']), max(test_data['x2']), 50)
y_p = (-classifier_p.W[1]/classifier_p.W[2])*x + (-classifier_p.W[0]/classifier_p.W[2])
a[0].plot(d1['x1'].values, d1['x2'].values, 'rx', label='class 1')
a[0].plot(d2['x1'].values, d2['x2'].values, 'bx', label='class 0')
a[0].plot(x,y_p,'g-', label='perceptron decision boundary')
a[0].set_title('train data')
a[1].plot(c1['x1'].values, c1['x2'].values, 'rx', label='class 1')
a[1].plot(c2['x1'].values, c2['x2'].values, 'bx', label='class 0')
a[1].plot(x,y_p,'g-', label='perceptron decision boundary')
a[1].set_title('test data')
plt.ylim(min(test_data['x2']), max(test_data['x2']))
plt.xlim(min(test_data['x1']), max(test_data['x2']))
plt.xlabel('x1')
plt.ylabel('x2')
a[0].legend()
a[1].legend()
plt.show()
```

```python
##logistic training and testing

classifier_p = logistic_regression.Logistic(2,1)
classifier_p.W = copy.deepcopy(train(20, 0, classifier_p, train_data))## train for 20 iterations
print('learnt parameters: ',classifier_p.W)
#test(classifier_p, train_data)
test(classifier_p, test_data)

d1 = train_data[train_data['class_label']==1]
d2 = train_data[train_data['class_label']==-1]

fig,a =  plt.subplots(1,2, figsize=(16,10))
x = np.linspace(min(test_data['x1']), max(test_data['x2']), 50)
y_p = (-classifier_p.W[1]/classifier_p.W[2])*x + (-classifier_p.W[0]/classifier_p.W[2])
a[0].plot(d1['x1'].values, d1['x2'].values, 'rx', label='class 1')
a[0].plot(d2['x1'].values, d2['x2'].values, 'bx', label='class 0')
a[0].plot(x,y_p,'g-', label='logistic decision boundary')
a[0].set_title('train data')
a[1].plot(c1['x1'].values, c1['x2'].values, 'rx', label='class 1')
a[1].plot(c2['x1'].values, c2['x2'].values, 'bx', label='class 0')
a[1].plot(x,y_p,'g-', label='logistic decision boundary')
a[1].set_title('test data')
plt.ylim(min(test_data['x2']), max(test_data['x2']))
plt.xlim(min(test_data['x1']), max(test_data['x2']))
plt.xlabel('x1')
plt.ylabel('x2')
a[0].legend()
a[1].legend()
plt.show()
```
